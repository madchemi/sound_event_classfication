```python
import os
import glob
import librosa
import numpy as np
import pandas as pd
import pickle
import resampy

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

max_pad_len = 174

def extract_feature(file_name):
    print('file name:', file_name)
    try:
        # librosa.load에서 res_type 인자 제거
        audio, sample_rate = librosa.load(file_name)
        mfccs = librosa.feature.mfcc(y=audio, sr=sample_rate, n_mfcc=40)
        pad_width = max_pad_len - mfccs.shape[1]
        mfccs = np.pad(mfccs, pad_width=((0,0), (0, pad_width)), mode='constant')
        print(mfccs.shape)
        return mfccs

    except Exception as e:
        print("Error encountered while parsing file:", file_name)
        print(e)
        return None

# 함수 실행
extract_feature('/content/drive/MyDrive/gist_공모전/archive-2/fold1/101415-3-0-2.wav')

    #np.pad(array, pad_width, mode = 'constant', **kwargs:
    # array: padding할 array, pad_width: 각 축에 padding에 사용할 숫자들(x,y), mode -> padding을 어떤 것으로 할지,
```

```python
#경로 설정 및

full_dataset_path = '/content/drive/MyDrive/gist_공모전/archive-2'
metadata = pd.read_csv("/content/drive/MyDrive/gist_공모전/archive-2/UrbanSound8K.csv")
features =[]

for index, row in metadata.iterrows():
   file_name = os.path.join(os.path.abspath(full_dataset_path),
                            'fold'+str(row["fold"])+'/',str(row["slice_file_name"]))


   class_label = row["classID"]
   data = extract_feature(file_name)
   features.append([data, class_label])

featuresdf = pd.DataFrame(features, columns = ['features','class_label'])

```
